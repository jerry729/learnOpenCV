```python
import cv2 as cv
import  numpy as np

img = cv.imread('./images/noisy.png', cv.IMREAD_GRAYSCALE)
blur = cv.GaussianBlur(img, (5,5), 0)

hist = cv.calcHist([blur], [0], None, [256], [0,256]) 
hist.ravel()
```

```python
hist_norm = hist.ravel() / hist.sum() # 每一个灰度的概率
print(hist_norm)
```

```python
Q = hist_norm.cumsum() #概率密度
print(Q)
```

```python
p1, p2 = np.hsplit(hist_norm, [1])
print(p1)
print(p2)
```
